```python
import matplotlib.pyplot as plt

features = [
'Final Score',
'Projects Score',
'Midterm Score',
'Assignments Avg',
'Participation Score',
'Quizzes Avg',
'Study Hours',
'Attendance'
]

importance = [
0.238943,
0.214584,
0.129374,
0.097821,
0.070506,
0.066749,
0.029962,
0.029296
]

plt.figure(figsize=(10,6))

plt.barh(
    features,
    importance
)

plt.title(
    "Feature Importance"
)

plt.xlabel(
    "Importance Score"
)

plt.tight_layout()

plt.savefig(
    "../frontend/static/feature_importance.png"
)

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = [
    "Low Risk",
    "Medium Risk",
    "High Risk"
]

sizes = [
    2584,
    1443,
    973
]

plt.figure(figsize=(6,6))

plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%"
)

plt.title(
    "Dropout Risk Distribution"
)

plt.savefig(
    "../frontend/static/dropout_risk.png"
)

plt.show()
```
